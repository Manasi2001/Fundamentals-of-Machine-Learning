### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn import preprocessing
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse 
```

### Getting our Data

```python
df = pd.read_csv(r'../Dataset/laptoppricing.csv', encoding = 'unicode_escape')
df
```

### Data Preprocessing

```python
# renaming a column
df.rename(columns = {'ï»¿Manufacturer':'Manufacturer'}, inplace = True)
```

```python
df
```

```python
# checking for null values
df.isnull().any()
```

```python
df.columns
```

```python
label_encoder = preprocessing.LabelEncoder()
df['Manufacturer'] = df['Manufacturer'].astype('|S')
df['Manufacturer'] = label_encoder.fit_transform(df['Manufacturer'])
```

```python
# checking vif
variables = df[['Manufacturer', 'IntelCore(i-)', 'IntelCoreGen',
       'processing speed(GHz)', 'Ram(gb)', 'HDD(gb)', 'SSD(gb)',
       'Graphics(gb)', 'ScreenSize(inch)']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns

vif
```

```python
# now, we'll drop columns which have vif>10
df = df.drop(['IntelCore(i-)','IntelCoreGen','processing speed(GHz)','HDD(gb)','ScreenSize(inch)'], axis=1)
df
```

### Data Visualization

```python
sb.pairplot(df)
```

```python
# using Pearson's correlation method
corr = df.corr(method='pearson')
sb.heatmap(corr, annot=True)
```

### Splitting Data for Training and Testing

```python
data = df.values
X,y = data[:,:-1], data[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)  # splitting in the ration 70:30
```

### Model

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
```

### Making Predictions and Checking Accuracy

```python
y_pred = regr.predict(X_test)
```

```python
print('R2 Score:',r2_score(y_test, y_pred))
```

```python
print('Root mean squared error:', (mse(y_test, y_pred))**(1/2))
```

# Predictions are 72.07% accurate.

### Visualizing the results

```python
fig = plt.figure()
m, b = np.polyfit(y_test, y_pred, 1)
plt.plot(y_test, m*y_test + b, color='red')
plt.scatter(y_test, y_pred, alpha=.5)
fig.suptitle('Actual vs Predicted Price', fontsize = 20)    # Plot heading 
plt.xlabel('Actual', fontsize = 18)                         # X-label
plt.ylabel('Predicted', fontsize = 16)                      # y-label
plt.show()
```

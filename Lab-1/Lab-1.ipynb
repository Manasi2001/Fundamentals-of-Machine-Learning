### *Performing basic operations on matrices (like addition, subtraction, multiplication) and display specific rows or columns of the matrix.*

```python
import numpy as np
  
# getting the input   
print("Enter matrix A:\n")
arr_1 = np.array([input().strip().split() for _ in range(3)], int) 
print("Enter matrix B:\n")
arr_2 = np.array([input().strip().split() for _ in range(3)], int)

# printing both the arrays
print("\nMatrix A:\n",arr_1)
print("\nMatrix B:\n",arr_1)

# adding the matrices
add_arrs = np.add(arr_1, arr_2) 
print("\nAfter addition:\n",add_arrs) 

# subtracting the matrices
sub_arrs = np.subtract(arr_1, arr_2) 
print("\nAfter subtraction:\n",sub_arrs) 

# multiplying the matrices
pro_arrs = np.multiply(arr_1, arr_2) 
print("\nAfter element by element multiplication:\n",pro_arrs) 

# transpose of matrices
print("\nAfter transpose matrix A:\n",arr_1.transpose())
print("\nAfter transpose matrix B:\n",arr_2.transpose())

# dot multiplication of matrices
dot_mul_arrs = np.dot(arr_1, arr_2)
print("\nAfter dot multiplication:\n",dot_mul_arrs)

# displaying a section of matrix A
print("\nA section of matrix A:\n",arr_1[1:,1:])
```
